```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import  tqdm

def make_plot(t,cnorm, lnorm,k,title,filename):
  plt.figure(figsize = (16,9))
  plt.plot(t,cnorm,label = "C")
  plt.plot(t,lnorm, label = "L2")
  plt.plot(t,[k]*len(t), label = 'Theory')
  plt.xlabel('t, sec', fontsize = 16)
  plt.ylabel('ratio', fontsize = 16)
  plt.title(title, fontsize = 16)
  plt.grid()
  plt.legend(fontsize = 16)
  plt.savefig(filename)
  plt.show()
  

#вычисление начального условия
def v0(a,b,r):
  if (r<b and r>a):
      return np.exp((-4*(2*r-a - b)**2)/((b-a)**2 - (2*r-a-b)**2))
  else:
      return 0

#Построение аналитического решения
def analytic_solution(r,t,c,a,b):
  ret = np.zeros((len(t), len(r)))
  for i in range(len(t)):
    for j in range(len(r)):
      ret[i,j] = v0(a,b,c*t[i]-r[j])
  return ret

#Решатель уравнения
def solver042(rmin,rmax,I,T,C,c,a,b,k):
  h = (rmax-rmin)/I
  r = np.zeros(I+2)
  r = np.linspace(rmin-0.5*h, rmax+0.5*h,I+2)
  tau = C*h/(c*k)
  u = np.zeros((T, I+2))
  for i in range(len(r)):
    u[0,i] = v0(a,b,-r[i])
    u[1,i] = v0(a,b,c*tau-r[i])
  for i in tqdm(range(1,T-1)):
    for j in range(2,len(r)-2):
      u[i+1,j] = 2*u[i,j]  - u[i-1,j]+ tau**2 * c**2 *( -5*u[i,j]/2 + 4*(u[i,j-1] + u[i,j+1])/3 - (u[i,j-2]+u[i,j+2])/12)/h**2
    u[i+1,1] = u[i+1,2]
    u[i+1,0] = u[i+1,1] 
    u[i+1,-2] = u[i+1,-3]
    u[i+1,-1] = u[i+1,-2]
  return u,r

def solver062(rmin,rmax,I,T,C,c,a,b,k):
  h = (rmax-rmin)/I
  r = np.zeros(I+2)
  r = np.linspace(rmin-0.5*h, rmax+0.5*h,I+2)
  tau = C*h/(c*k)
  u = np.zeros((T, I+2))
  for i in range(len(r)):
    u[0,i] = v0(a,b,-r[i])
    u[1,i] = v0(a,b,c*tau-r[i])
  for i in tqdm(range(1,T-1)):
    for j in range(3,len(r)-3):
      u[i+1,j] = 2*u[i,j]  - u[i-1,j]+ tau**2 * c**2 *( -49*u[i,j]/18 + 3*(u[i,j-1] + u[i,j+1])/2 - 3*(u[i,j-2]+u[i,j+2])/20 + (u[i,j-3]+u[i,j+3])/90 )/h**2
    u[i+1,2] = u[i+1,3]
    u[i+1,1] = u[i+1,2]
    u[i+1,0] = u[i+1,1]
    u[i+1,-3] = u[i+1,-4] 
    u[i+1,-2] = u[i+1,-3]
    u[i+1,-1] = u[i+1,-2]
  return u,r
```

```python
rmin = 0
rmax = 1.8
c = 1.5
a = -0.7
b = -0.1
C = 0.8

T_1 = 50
T_2 = 50*9
T_3 = 50*81 

I_1 = 100
I_2 = 300
I_3 = 900


u_11, r1 = solver042(rmin,rmax,I_1,T_1,C,c,a,b,1)
t1 = [ C*(r1[1] - r1[0])*i/(c) for i in range(T_1)]

u_a = analytic_solution(r1,t1,c,a,b)

u_12, _ = solver042(rmin,rmax,I_2,T_2,C,c,a,b,3)
u_13, _ = solver042(rmin,rmax,I_3,T_3,C,c,a,b,9)

u_21, _ = solver062(rmin,rmax,I_1,T_1,C,c,a,b,1)
u_22, _ = solver062(rmin,rmax,I_2,T_1*3**3,C,c,a,b,9)
u_23, _ = solver062(rmin,rmax,I_3,T_1*3**6,C,c,a,b,81)
```

```python
def get_u_indx(u,tindx,rindx):
  newu = np.zeros((len(tindx), len(rindx)))
  for i in range(len(tindx)):
    for j in range(len(rindx)):
      newu[i,j] = u[tindx[i], rindx[j]]
  return newu

tindx_12 = [i for i in range(0,len(u_12),9)]
tindx_13 = [i for i in range(0,len(u_13),81)]

tindx_22 = [i for i in range(0,len(u_22),27)]
tindx_23 = [i for i in range(0,len(u_23),3**6)]

rindx_2 = [i for i in range(2,len(u_12[0]),3)]
rindx_3 = [i for i in range(5,len(u_13[0]),9)]

newu11 = u_11[:,1:-1]
newua = u_a[:,1:-1]
newu12 = get_u_indx(u_12,tindx_12,rindx_2)
newu13 = get_u_indx(u_13,tindx_13,rindx_3)

newu21 = u_21[:,1:-1]
newu22 = get_u_indx(u_22,tindx_22,rindx_2)
newu23 = get_u_indx(u_23,tindx_23,rindx_3)
```

```python
cnorm1 = []
lnorm1 = []

cnorma1 = []
lnorma1 = []

cnorm2 = []
lnorm2 = []

cnorma2 = []
lnorma2 = []

for i in range(newu11.shape[0]):
  cnorm1.append(np.max(np.abs(newu11[i] - newu12[i]))/np.max(np.abs(newu12[i] - newu13[i])))
  lnorm1.append(np.sqrt(np.sum((newu11[i] - newu12[i])**2))/np.sqrt(np.sum((newu12[i] - newu13[i])**2)))
  
  if (np.isclose(np.max(np.abs(newu11[i] - newua[i])),0,atol=1e-7) or np.isclose(np.max(np.abs(newu12[i] - newua[i])),0,atol=1e-7) ) :
    cnorma1.append(81)
  else:
    cnorma1.append(np.max(np.abs(newu11[i] - newua[i]))/np.max(np.abs(newu12[i] - newua[i])))

  if(np.isclose(np.sqrt(np.sum((newu11[i] - newua[i])**2)),0,atol = 1e-7) or np.isclose(np.sqrt(np.sum((newu12[i] - newua[i])**2)),0,atol = 1e-7)):
    lnorma1.append(81)
  else:
    lnorma1.append(np.sqrt(np.sum((newu11[i] - newua[i])**2))/np.sqrt(np.sum((newu12[i] - newua[i])**2)))

  cnorm2.append(np.max(np.abs(newu21[i] - newu22[i]))/np.max(np.abs(newu22[i] - newu23[i])))
  lnorm2.append(np.sqrt(np.sum((newu21[i] - newu22[i])**2))/np.sqrt(np.sum((newu22[i] - newu23[i])**2)))
  
  if (np.isclose(np.max(np.abs(newu21[i] - newua[i])),0,atol=1e-7) or np.isclose(np.max(np.abs(newu22[i] - newua[i])),0,atol=1e-7) ) :
    cnorma2.append(3**6)
  else:
    cnorma2.append(np.max(np.abs(newu21[i] - newua[i]))/np.max(np.abs(newu22[i] - newua[i])))

  if(np.isclose(np.sqrt(np.sum((newu21[i] - newua[i])**2)),0,atol = 1e-7) or np.isclose(np.sqrt(np.sum((newu22[i] - newua[i])**2)),0,atol = 1e-7)):
    lnorma2.append(3**6)
  else:
    lnorma2.append(np.sqrt(np.sum((newu21[i] - newua[i])**2))/np.sqrt(np.sum((newu22[i] - newua[i])**2)))
```

```python
make_plot(t1,cnorm1,lnorm1,81,"Уменьшение шага в 3 раза и в 9 раз для о42","o42.png")
```

```python
make_plot(t1,cnorma1,lnorma1,81,"Сравнение с аналитическим решением для o42","o42a.png")
```

```python
make_plot(t1,cnorm2,lnorm2,729,"Уменьшение шага в 3 раза и в 9 раз для о62","o62.png")
```

```python
make_plot(t1,cnorma2,lnorma2,729,"Сравнение с аналитическим решением для о62","o62a.png")
```
